```python
import glob 
import pandas as pd
import numpy as np
from datetime import datetime
from pandas import ExcelWriter
import xlrd
import matplotlib.pyplot as plt
```

```python
Const_pyr1 = 1000/70.6      #(W/m2)/(mV)
Const_pyr2 = 1000/8.81      #(W/m2)/(mV)

files = glob.glob('*.xls')
sensors = pd.DataFrame()


for i in files:
    try:
        lectura1 = pd.read_excel(i, sheet_name='Resumen', skiprows = 1)
        sensors = sensors.append(lectura1)
    except xlrd.XLRDError:
        print('Archivo '+i+' sin hoja Resumen')
    
for i in files:
    try:
        lectura2 = pd.read_excel(i, sheet_name='Resumen 2', skiprows = 1)
        sensors = sensors.append(lectura2)
    except xlrd.XLRDError:
        print('Archivo '+i+' sin hoja Resumen 2')
```

```python
sensors = sensors.rename(columns={' ':'TS_date',' .1':'TS_hour','"mV"':'CH1_mV','"mV".1':'CH2_mV'})
sensors['CH1_mV'] = pd.to_numeric(sensors.CH1_mV, errors='coerce').fillna(0)
sensors['CH2_mV'] = pd.to_numeric(sensors.CH2_mV, errors='coerce').fillna(0)

sensors['CH1_W/m2'] = float(Const_pyr1)*sensors["CH1_mV"]
sensors['CH2_W/m2'] = float(Const_pyr2)*sensors["CH2_mV"]

sensors['TS_hour'] = sensors['TS_hour'].str.replace('a. m.','AM')
sensors['TS_hour'] = sensors['TS_hour'].str.replace('p. m.','AM')
```

```python
sensors.index = pd.to_datetime(sensors['TS_date'] + ' ' + sensors['TS_hour'], format ="%d/%m/%Y %I:%M:%S %p")
sensors[(sensors.index.year > 2016)]
sensors[(sensors.index.year < 2017)]
```

```python
sensors = sensors.resample("5T").mean()
sensors
```

```python
sensors.plot()
```

```python
sensors.insert(0, 'TS', sensors.index)
sensors.to_csv('Sensors.csv',index=False)
```
